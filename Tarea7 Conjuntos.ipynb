Ejercicio 1
Dado un número entero introducido por teclado, guarda sus divisores en un conjunto y muéstralo.

```python
n = int(input("Introduzca un número entero: "))
conjunto = set()
for i in range(2, n//2):
  if n % i == 0:
    conjunto.add(i)

print(conjunto)
```

Ejercicio 2
Crea un programa que dado un conjunto, nos devuelva su mínimo. Debes hacerlo sin recurrir a la función
min().

```python
conjunto = set()
n = int(input("Introduzca el número de elementos que contendrá el conjunto: "))
print("Introduzca los elementos de su conjunto")
for _ in range(n):
  numero = int(input())
  conjunto.add(numero)
minimo = 1000000
for elemento in conjunto:
  if elemento < minimo:
    minimo = elemento

print("El elemento mínimo de {} es {}.".format(conjunto,minimo))
```

Ejercicio 3
Dada una frase introducida por teclado, guarda en un conjunto todas las palabras que empiecen por la letra
indicada por el usuario.

```python
frase = input("Introduzca una frase: ")
palabra = input("Introduzca una palabra: ")
conjunto = set()
frase = frase.lower()
palabra = palabra.lower()
frase = frase.split(" ")
for s in frase:
  if s[0] == palabra:
    conjunto.add(s)
print(conjunto)
```

Ejercicio 4
Dado un conjunto, crea un programa que nos devuelva el caracter con mayor valor ASCII. Debes hacerlo sin
recurrir a la función max().

```python
conjunto = set()
n = int(input("Introduzca el número de elementos que contendrá el conjunto: "))
print("Introduzca los elementos de su conjunto")
for _ in range(n):
  letra = input()
  conjunto.add(letra)
max = 0
for c in conjunto:
  if ord(c) > max:
    max = ord(c)
print("El caracter con mayor valor ASCII es:")
print(chr(max))
```

Ejercicio 5
Dado un conjunto, crea un programa que nos devuelva el caracter con menor valor ASCII. Debes hacerlo sin recurrir a la función min().

```python
conjunto = set()
n = int(input("Introduzca el número de elementos que contendrá el conjunto: "))
print("Introduzca los elementos de su conjunto")
for _ in range(n):
  letra = input()
  conjunto.add(letra)
min = 255
for c in conjunto:
  if ord(c) < min:
    min = ord(c)
print("El caracter con menor valor ASCII es:")
print(chr(min))
```

Ejercicio 6
Dada una frase introducida por teclado, guarda en un conjunto todas las palabras que contengan la letra indicada por el usuario.

```python
frase = input("Introduzca una frase: ")
letra = input("Introduzca una laetra: ")
frase = frase.lower()
letra = letra.lower()
frase = frase.split(" ")
conjunto = set()
for word in frase:
  if word.count(letra) > 0:
    conjunto.add(word)
print(conjunto)
```

Ejercicio 7
Dada una frase introducida por teclado, guarda en un conjunto la primera letra de cada palabra sin hacer uso del método .split().


```python
frase = input("Introduzca una frase: ")
frase = frase.lower()
frase = frase.strip()
conjunto = set()
for i in range(len(frase)):
  conjunto.add(frase[0])
  if frase[i] == " ":
    conjunto.add(frase[i+1])
print(conjunto)
```

Ejercicio 8
Dada una frase introducida por teclado, guarda en un conjunto todas las palabras con longitud par.

```python
frase = input("Introduzca una frase: ")
frase = frase.lower()
frase = frase.split()
conjunto = set()
for palabra in frase:
  if len(palabra)%2 == 0:
    conjunto.add(palabra)
print(conjunto)
```

Ejercicio 9
Dada una frase introducida por teclado, guarda en un conjunto todas las palabras que acaben por la letra indicada por el usuario.

```python
frase = input("Introduzca una frase: ")
frase = frase.lower()
frase = frase.split()
letra = input("Introduzca una letra: ")
letra = letra.lower()
conjunto = set()
for s in frase:
  if s[len(s)-1] == letra:
    conjunto.add(s)
print(conjunto)
```

Ejercicio 10
Dada una frase introducida por teclado, guarda en un conjunto todas las palabras palíndromas.

```python
frase = input("Introduzca una frase: ")
frase = frase.lower()
frase = frase.split()
conjunto = set()
for palabra in frase:
  var = True
  for i in range(len(palabra)//2):
    if palabra[i] != palabra[len(palabra)-1]:
      var = False
      break
  if var == True:
    conjunto.add(palabra)
print(conjunto)
```
