```python
class Date:

  def __init__(self, day=1, month=1, year=1):
    self.day = day
    self.month = month
    self.year = year
    if not self.validDate() :
      print("¡¡¡La fecha introducida no es válida!!!")

  def _fill_zeros(self, number, digits):
     return '0'*(digits - len(str(number))) + str(number)

  def isLeap(self):
    if (self.year % 4 == 0 and self.year % 100 and self.year % 400):
      return True
    else:
      return False

  def totalMonthDays(self):
    month = self.month
    year = self.year
    if month == 2:
      if self.isLeap():
        total_days = 29
      else:
        total_days = 28
    else:
      if (month == 1 or month == 3 or month == 5 or month == 7 or month == 8 or
         month == 10 or month == 12):
        total_days = 31
      else:
        total_days = 30
    return total_days

  def validDate(self):
    if self.month > 12 or self.month <= 0 :
      return False
    else:
      total_days = self.totalMonthDays()
      if self.day > total_days:
        return False
      else:
        return True

  @property
  def monthName(self):
    months = {
        1 : 'January',
        2 : 'February',
        3 : 'March',
        4 : 'April',
        5 : 'May',
        6 : 'June',
        7 : 'July',
        8 : 'August',
        9 : 'September',
        10 : 'October',
        11 : 'November',
        12 : 'December'}
    return months.get(self.month)

  @staticmethod
  def areEqual(date1, date2):
    if (date1.day == date2.day and date1.month == date2.month and
       date1.year == date2.year):
      return True
    else:
      return False

  @staticmethod
  def isLater(date1, date2):
    if (date1.year < date2.year):
      return False
    else:
      if date1.year > date2.year:
        return True
      else:
        if date1.month < date2.month:
          return False
        else:
          if date1.month > date2.month:
            return True
          else:
            if date1.day <= date2.day:
              return False
            else:
              return True

  @staticmethod
  def isPrevious(date1, date2):
    return not(Date.isLater(date1, date2) or Date.areEqual(date1, date2))

  @classmethod
  def fristDayOfTheYear(cls, year):
    return cls(1, 1, year)

  @classmethod
  def lastDayOfTheYear(cls, year):
    return cls(31, 12, year)

  def plusDay(self):
    if self.day + 1 <= self.totalMonthDays():
      self.day += 1
    else:
      if self.month == 12:
        self.day = 1
        self.month = 1
        self.year += 1
      else:
        self.day = 1
        self.month +=1

  def minusDay(self):
    if self.day - 1 > 0 :
      self.day -= 1
    else:
      if self.month ==1:
        self.day = 31
        self.month = 12
        self.year -= 1
      else:
        self.month -= 1
        self.day = self.totalMonthDays()

  @classmethod
  def copy(cls, original_date):
    return cls(original_date.day, original_date.month, original_date.year)

  @staticmethod
  def difference(date1, date2):
    sum = 0
    if Date.areEqual(date1, date2):
      return sum
    else:
      if Date.isPrevious(date1, date2):
        start_date = Date.copy(date1)
        finish_date = Date.copy(date2)
      else:
        start_date = Date.copy(date2)
        finish_date = Date.copy(date1)
      while not Date.areEqual(start_date, finish_date):
        start_date.plusDay()
        sum += 1
      return sum

  @classmethod
  def randomDate(cls):
    year = random.randrange(1,3000)
    month = random.randrange(1,13)
    day = random.randrange(1, 31)
    if cls(day, month, year).validDate():
      return cls(day, month, year)
    else:
      while not cls(day, month, year).validDate():
        year = random.randrange(1,3000)
        month = random.randrange(1,13)
        day = random.randrange(1, 31)
      return cls(day, month, year)

  @classmethod
  def toDate(cls, stringDate):
    day = int(stringDate.split()[0])
    year = int(stringDate.split()[-1])
    months = {
        'January':1,
        'February':2,
        'March':3,
        'April':4,
        'May':5,
        'June':6,
        'July':7,
        'August':8,
        'September':9,
        'October':10,
        'November':11,
        'December':12}
    return cls(day, months.get(stringDate.split()[-2]), year)

  def __str__(self):
    formatted_day = self._fill_zeros(self.day, 2)
    formatted_month = self._fill_zeros(self.month, 2)
    formatted_year = self._fill_zeros(self.year, 4)
    return ('{}/{}/{}'.format(formatted_day, formatted_month, formatted_year))
```
