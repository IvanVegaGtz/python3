Ejercicio 3
Dada una lista de palabras, quédate con filter() con las que tengan más vocales que consonantes. Necesitarás una función que devuelva si una palabra tiene más vocales que consonantes.

```python
palabras = ["a", "all", "ala", "hola", "perro", "ave", "killers"]
def masvocales(word):
  vocal = 0
  consonante = 0
  word = word.lower()
  for letter in word:
    if letter=='a' or letter=='e' or letter=='i' or letter=='o' or letter=='u':
      vocal += 1
    else:
      consonante += 1
  if vocal > consonante:
    return True
list(filter(masvocales, palabras))
```

Ejercicio 4
Dada una lista de números enteros, quédate con filter() con los que tengan más de 5 divisores. Necesitarás
una función que devuelva el número de divisores de un número dado.

```python
numeros = [1000, 425, 81, 17, 51, 100]
def cardnum(num):
  card = 1
  for i in range(1, num//2 + 1):
    if num%i == 0:
      card +=1
  return card
cardnum(17)
list(filter(lambda x: cardnum(x) > 5, numeros))
```

Ejercicio 5
Dada una lista de palabras, quédate con reduce() con la palabra más larga. Necesitarás una función que
compare dos palabras y devuelva la que tenga mayor longitud.

```python
from functools import reduce
palabras = ["perro", "gato", "negro", "milka", "ornitorrinco", "jesse", "ferrcoarrill"]
def compare_words(word1, word2):
  if len(word1) >= len(word2):
    return word1
  return word2
reduce(compare_words, palabras )
```

Ejercicio 6
Dada una lista de palabras, calcula el número de vocales de cada una con map().

```python
palabras = ["perro", "gato", "negro", "milka", "ornitorrinco", "jesse", "ferrcoarrill"]
def count_vocales(word):
  cont = 0
  word = word.lower()
  for s in word:
    if s=='a' or s=='e' or s=='i' or s=='o' or s=='u':
      cont += 1
  return cont
list(map(count_vocales,palabras))
```

Ejercicio 7
Dada una lista de palabras, quédate con reduce() y map() con la palabra con más consonantes. Necesitarás
una función que cuente el número de consonantes de una palabra y otra que dados dos números, devuelva el
mayor.

```python
from functools import reduce
def count_consonantes(word):
  cont = 0
  word = word.lower()
  for s in word:
    if s!='a' and s!='e' and s!='i' and s!='o' and s!='u':
      cont += 1
  return cont

def mayor(x,y):
  if x >= y:
    return x
  else:
    return y

palabras = ["perro", "gato", "negro", "milka", "ornitorrinco", "jesse", "ferrcoarrill"]
reduce(mayor,list(map(count_consonantes, palabras)))
```
