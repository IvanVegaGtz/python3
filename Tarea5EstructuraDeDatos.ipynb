Ejercicio 1
Crea un programa que lea una secuencia de caracteres, guarde cada caracter en una posición de una lista y
finalmente muestre la secuencia invertida.

```python
n = int(input("Introduzca el número de caracteres que desea ingresar: "))
lista = []
for i in range(n):
  lista.append(input())
lista.reverse()
print(lista)

```

Ejercicio 2
Crea un programa que lea dos strings de la misma longitud, los guarde intercalados en una lista. Por último,
mostrar el contenido de la lista.
Por ejemplo, si introducimos hola caracola y adios marieta, debería mostrarse haodleao sc
amraarcioeltaa.

```python
palabra1 = input("Introduzca la primer palabra: ")
palabra2 = input("Introduczca la segunda palabra: ")
n = len(palabra1)
m = len(palabra2)
cont1, cont2 = 0,0
lista = []
if n == m:
  for i in range(2*n):
    if i%2 == 0:
      lista.append(palabra1[cont1])
      cont1 += 1
    else:
      lista.append(palabra2[cont2])
      cont2 += 1
else:
  print("Las palabras deben tener la misma longitud")
print(lista)
```

Ejercicio 3
Crea un programa que lea un string y guarde en una lista todas las consonantes.

```python
palabra = input("Introduzca un string: ")
S = palabra.lower()
lista = []
for i in range(len(S)):
  if S[i] != 'a' and S[i] != 'e' and S[i] != 'i' and S[i] != 'o' and S[i] != 'u':
    lista.append(palabra[i])
print(lista)
```

Ejercicio 4
Crea un programa que lea una palabra, la guarde en una lista y compruebe si se trata de un palíndromo.

```python
palabra = input("Introduzca una palabra: ")
lista= []
es_palindroma = True
for j in range(len(palabra)):
  lista.append(palabra[j])
for i in range (len(palabra)):
  if lista[i] != lista[len(palabra)-i-1]:
    es_palindroma = False
    break
if (es_palindroma == True):
  print("La palabra es palindroma")
else:
  print("La palabra no es palindroma")
```

Ejercicio 5
Crea un programa que lea una matriz 3 x 3 y devuelva el máximo de cada fila.

```python
M = []
maximos = []
for i in range(3):
  M.append([])
  for j in range(3):
    M[i].append(int(input()))

for i in range(3):
  max = M[i][0]
  for j in range(2):
      if max <= M[i][j+1]:
        max = M[i][j+1]
  maximos.append(max) 

for i in range(3):
  print("El valor máximo de la fila {} es {}".format(i+1,maximos[i]))
```

Ejercicio 8
Crear un programa determina si la matriz introducida manualmente (tanto sus dimensiones como los elementos) se trata de la matriz identidad. Recuerda que la matriz identidad debe ser una matriz cuadrada.

```python
n = int(input("Introduzca la dimensión de la matriz: "))
M = []
identidad = []
es_identidad = True

for i in range(n):
  M.append([])
  identidad.append([])
  for j in range(n):
    M[i].append(int(input()))
    if i == j:
      identidad[i].append(1)
    else:
      identidad[i].append(0)

for i in range(n):
  for j in range(n):
    if M[i][j] != identidad[i][j]:
      es_identidad = False
      break

if es_identidad == True:
  print("La matriz que introdujo es la matriz identidad")
else:
  print("La matriz que introdujo no es la matriz identidad")
```

Ejercicio 9
Realiza un programa que calcule la matriz transpuesta

```python
n = int(input("Introduzca el número de filas: "))
m = int(input("Introduzca el número de columnas: "))
M = []

for i in range(n):
  M.append([])
  for j in range(m):
    M[i].append(int(input()))

trans = []

for i in range(m):
  trans.append([])
  for j in range(n):
    trans[i].append(M[j][i])
    print(trans[i][j], end=" ")
  print("")
```

Ejercicio 10
Crea un programa que pida al usuario la dimensión y cree la matriz identidad de orden correspondiente con
numpy.

```python
import numpy as np
n = int(input("Introduzca la dimensión de la matriz: "))
I = np.zeros((n,n))
for i in range(n):
  for j in range(n):
    if i == j:
      I[i][j] = 1
print(I)
```
